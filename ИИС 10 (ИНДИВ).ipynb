# 1.2 Загрузка данных и разделение на матрицу признаков и целевую переменную

```python
data = pd.read_csv("ice_cream_selling.csv", delimiter=',')
data.head()
```

```python
X = data.iloc[:, 0].values
y = data.iloc[:, -1].values
print("Матрица признаков"); print(X)
print("Зависимая переменная"); print(y)
```

# 1.3 Разделение выборки на тренировочную и тестовую

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

# 1.4 Обучение линейной модели



```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
X = X.reshape(-1, 1)
lin_reg.fit(X, y)
```

# 1.5 Обучение полиноминальной модели

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

# 1.6 Предсказание, обработка и визуализация результатов




```python
data.describe()
```

```python
sample_data = [[10.12]]
y_pred_lin = lin_reg.predict(sample_data)
y_pred_poly = lin_reg_2.predict(poly_reg.fit_transform(sample_data))
print(y_pred_lin, y_pred_poly)
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.xlabel("Distance_to_center")
plt.ylabel("Price")
plt.title("Truth or bluff (Linear Regression)")
plt.show()
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color='blue')
plt.xlabel("Square feet")
plt.ylabel("Price")
plt.title("Truth or bluff (Linear Regression)")
plt.show()
```

```python
poly_reg = PolynomialFeatures(degree=3)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)

plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color='blue')
plt.xlabel("Square feet")
plt.ylabel("Price")
plt.title("Truth or bluff (Polynomial Regression, n=3)")
plt.show()
```
