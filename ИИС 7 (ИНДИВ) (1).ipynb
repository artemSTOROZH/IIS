# 2. Загрузка данных

```python
data = pd.read_csv('heart_failure_clinical_records_dataset.csv', delimiter=',', header=0)
data.head()
```

```python
data.shape
```

```python
import random

# Создание пропущенных значений в датасете
for i in range(data.shape[0]):
  prob = random.random()
  if prob < 0.15:
    for j in range(13):
      data.iloc[i, j] = None
data.info()
```

# 3. Разделение выборки на тренировочную и тестовую

```python
X = data.iloc[:, :-1]
y = data.iloc[:, -1]
print("Матрица признаков"); print(X)
print("Зависимая переменная"); print(y)
```

# 4. Обработка отсутствующих значений

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.NaN, strategy='mean')
imputer = imputer.fit(X.iloc[:, [0, 2, 4, 6, 7, 8, 11]])
X.iloc[:, [0, 2, 4, 6, 7, 8, 11]] = imputer.transform(X.iloc[:, [0, 2, 4, 6, 7, 8, 11]])
print(X)
```

# 5. Обработка значений зависимой переменной

```python
from sklearn.preprocessing import LabelEncoder
label_encoder_X_cat = LabelEncoder()
label_encoder_y = LabelEncoder()

print("Зависимая переменная до обработки")
print(y)
y = label_encoder_y.fit_transform(y)
print("Зависимая переменная после обработки")
print(y)
```

# 6. Обработка значений категориальных признаков

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
X = encoder.fit_transform(X.iloc[:,[1, 3, 5, 9, 10]])
print("Перекодировка категориального признака")
print(X)
```
